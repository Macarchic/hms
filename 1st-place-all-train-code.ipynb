# Hello!

Below you can find a outline of how to reproduce my solution for the HMS - Harmful Brain Activity Classification competition.

# HARDWARE: (The following specs were used to create the original solution)
- OS  : Ubuntu 20.04 LTS
- CPU : Intel(R) Core(TM) i9-10980XE CPU @ 3.00GHz
- GPU : 2 x NVIDIA RTX3090

# SOFTWARE 
- python packages are detailed separately in `requirements.txt`
  - jax / jaxlib depends on the cuda / cudnn version, so you need to find the appropriate version from [this link](https://storage.googleapis.com/jax-releases/jax_cuda_releases.html), and change the version in requirements.txt.
- Python 3.8.10
- CUDA 11.3
- cudnn 8.2.0
- nvidia drivers v.470.239.06

# DATA SETUP 
- assumes the [Kaggle API](https://github.com/Kaggle/kaggle-api) is installed
- below are the shell commands used in each step, as run from the top level directory
```
mkdir -p dataset
cd dataset
kaggle competitions download -c hms-harmful-brain-activity-classification
unzip hms-harmful-brain-activity-classification.zip
cd ..
```

# OVERVIEW for model build
Our model is an ensemble of 30 models trained using four different methods.
- We describe three approaches for training: 
  - A. One where all models are trained collectively 
  - B. Another where the training is conducted individually in four parts
  - C. And, we have prepared command to train best cv model

## First : COMMON DATA PROCESSING 
Preprocess train.csv. You need to run this command for training.
```
python3 generate_groupkfold_csv.py
```

## A. TRAIN ALL MODEL 
This will take about 7 days (or more).
```
source train_all.sh
```

## B. Train Specific MODEL
### yamash models
If you train a specific model, please read a shell script.
```
cd yamash
source prepare_data_train.sh
```
  
### suguuuuu models
If you train a specific model, please read a shell script.
```
cd suguuuuu
source prepare_data_train.sh
```

### muku models
If you train a specific model, please read a shell script.
```
cd muku
source prepare_data_train.sh
```
If you want to check the operation simply in debug mode, please give the -d option.
```
source prepare_data_train.sh -d
```

### kfuji models
If you train a specific model, please read a shell script.
```
cd kfuji
source prepare_data_train.sh
```

## C. Train Best CV model
The best cv is No.19 (written in bottom list) in our model. If you want to train the model, please run below command.
```
cd muku
python3 prepare_data.py
python3 train_v046_2_to_v062.py v057
```
